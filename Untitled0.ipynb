# Distribuição Binomial

A distribuição binomial é utilizada para calcular a probabilidade de ocorrência de um determinado número de sucessos em `x` eventos (ou ensaios) independentes, ou seja, quando cada aferição de ensaio possui apenas dois resultados possíveis.

**OU SEJA,** quando se deseja encontrar a probabilidade de quantidade de ocorrências de um evento com duas possibilidades.

## Exemplo de Aplicação da Hipótese Binomial

Um exemplo simples onde se pode aplicar a hipótese Binomial é para ensaios de resistência mecânica, como por exemplo, ensaios de resistência à compressão. Nesse caso, há uma resistência característica especificada em projeto, e cada corpo de prova pode ser considerado como um experimento independente. Os valores encontrados em cada ensaio possuem apenas dois resultados possíveis:

*   **Sucesso:** se o valor da resistência for maior ou igual à resistência de projeto.
*   **Fracasso:** se a resistência for menor do que a resistência especificada.


A seguir, um exemplo programado:

```python
import math

#Em uma obra com 10 corpos de prova para ensaio de resistência a compressão, qual a probabilidade de 1 dar abaixo da resistência esperada.

# número total de corpos de prova
n = 10

# número de cp que podem dar abaixo da resistência esperada:
k = 1

# p= probabilidade de falha. Nesse caso a probabilidade de falha de um corpo de prova precisa ser estimada.
#Em estudos reais esse dado deve vir de dados observados na literatura ou referências de ensaios.
p = 0.01

# fórmula da distribuição binomial
probabilidade = (math.comb(n, k)) * (p**k) * ((1-p)**(n-k))

print("Probabilidade de exatamente", k, "corpos de prova falharem:", probabilidade)
```

# Limitações da Distribuição Binomial em Ensaios de Durabilidade

Porém, essa forma de resposta não é esperada em **ensaios de durabilidade.**

Os ensaios de durabilidade oferecem valores que são sempre comparados dentro de uma faixa.

Os resultados obtidos são variáveis contínuas, que podem assumir uma ampla faixa de valores numéricos, como por exemplo, os ensaios de resistividade, de carbonatação, de cloretos.

Assim, não existem apenas dois resultados possíveis para cada ensaio, mas sim muitos valores diferentes.

Por esse motivo, a distribuição binomial não é adequada para tratar esses dados. O mais apropriado para análises desses resultados é utilizar distribuições contínuas, como a distribuição normal.

Vamos ver um exemplo de forma bem detalhada.

# Distribuição Nominal

A distribuição normal é uma distribuição de probabilidade contínua que descreve comportamentos onde os valores tendem a se concentrar em torno de um valor médio, e apresentam uma dispersão de forma simétrica em relação a essa média, que é o desvio padrão.

Em muitos fenômenos naturais e de ensaios laboratoriais os resultados tendem a seguir esse comportamento, fazendo com que a distribuição normal seja muito mais utilizada na análise estatística.

Nos ensaios de durabilidade do concreto (inclusive utilizados na minha pesquis), como resistividade elétrica, coeficiente de difusão ou teor de cloretos, os resultados são valores numéricos contínuos que podem variar dentro de uma faixa de medidas e permitem cálculos como média e desvio padrão.

Vamos ver um exemplo de forma bem detalhada.






## Exemplo de Aplicação da Hipótese Nominal

Em um ensaio de resistividade elétrica realizado em 10 corpos de prova de concreto de mesmo traço, são obtidas 8 leituras de resistividade para cada corpo de prova, distribuídas em torno de um valor médio.

As medidas variam devido à heterogeneidade do concreto (umidade, porosidade e posicionamento dos eletrodos).
Os valores do ensaio se concentram em torno de uma média e apresentam dispersão aproximadamente simétrica, sendo compatível com a distribuição normal.

De acordo com o procedimento de ensaio, calcula-se inicialmente a média das leituras e o desvio padrão, que representa a variabilidade dos resultados.

As leituras que se afastam significativamente do conjunto de dados podem ser descartadas com base em critérios estatísticos associados ao desvio padrão, sendo então calculada uma nova média representativa da resistividade do concreto.

Assim, a distribuição normal pode ser utilizada para analisar a variabilidade dos resultados experimentais e auxiliar na interpretação estatística das medições de resistividade do concreto.

A seguir, um exemplo programado para este ensaio:

```python
# Trazendo bibliotecas de 3 ferramentas básicas necessarias:
import numpy as np                  # Para fazer as contas de matemática
import matplotlib.pyplot as plt     # Para desenhar o gráfico
from scipy.stats import norm        # Para fazer a linha da curva normal

# 1. Os valores lidos de resistividade superficial nos corpos de prova:
dados_resistividade = [23.5, 22.1, 22.8, 21.7, 22.5, 21.9, 22.8, 21.2, 22.4, 22.2]

# 2. Calculo de Média e Desvio Padrão
media = np.mean(dados_resistividade)
desvio_padrao = np.std(dados_resistividade, ddof=1) # ddof=1 indica que é para uma amostra

# 3. Mostrando os resultados para esse corpo de prova:
print("--- RESULTADOS DA ANÁLISE ---")
print("Média:", round(media, 2), "kΩ.cm")
print("Desvio Padrão:", round(desvio_padrao, 2), "kΩ.cm")
print("-----------------------------")

# 4. Criando o Gráfico (Histograma)
# bins=5 divide as barras, edgecolor coloca uma linha preta em volta delas
plt.hist(dados_resistividade, bins=5, density=True, color='lightblue', edgecolor='black')

# 5. Desenhando a linha vermelha da Curva Normal
# Criamos vários pontinhos no eixo X (do valor 19 até o 25) para a linha ficar suave
eixo_x = np.linspace(19, 25, 100)

# Calculamos o formato do "sino" (a curva normal) usando a nossa média e desvio
curva = norm.pdf(eixo_x, media, desvio_padrao)

# Desenhamos a linha vermelha no gráfico
plt.plot(eixo_x, curva, color='red', linewidth=2)

# 6. Colocando os títulos para o gráfico ficar bonito e fácil de entender
plt.title("Gráfico de Resistividade dos Corpos de Prova")
plt.xlabel("Resistividade (kΩ.cm)")
plt.ylabel("Frequência")

# 7. Comando final para o gráfico aparecer na tela
plt.show()
```

Uma vez gerado o gráfico de resistividade desse corpo de prova e existindo outros 10 corpos de prova (com as mesmas caracterisitcas de traço) para realizar o mesmo ensaiocom podemos estimar a probabilidade de encontrar a resistividade média para eles tambem.
